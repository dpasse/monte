```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/mlb/eeyores-853208662-batters.csv')
```

```python
## totals by day,

df['MONTH'] = df.DATE.map(lambda date: date.split('/')[0]).astype(int)
df['DAY'] = df.DATE.map(lambda date: date.split('/')[1]).astype(int)

columns = ['MONTH', 'DAY', 'AB', 'R', 'H', 'HR', 'RBI', 'SB']
df[columns].groupby(['MONTH', 'DAY']).sum()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def moneyline_to_percent(moneyline: int) -> float:
    return round(
        (moneyline / (moneyline - 100) if moneyline < 0 else 100 / (moneyline + 100)) * 100,
        2
    )

moneyline_to_percent(109)
```

```python
plt.figure(figsize=(20, 10))

plt.title('Moneyline vs Percent')

x = np.linspace(-1000, -100, 1000)
y = [ moneyline_to_percent(v) for v in x ]
plt.plot(x, y, c='blue')

x = np.linspace(100, 1000, 1000)
y = [ moneyline_to_percent(v) for v in x ]
plt.plot(x, y, c='blue')

plt.xlabel('Moneyline')
plt.ylabel('Percent')

plt.show()
```

```python
def percent_to_moneyline(percentage: float) -> int:
    offset = (100 - percentage)
    return int(
        round(
            (percentage / offset) * -100 if percentage > 50 else (offset / percentage) * 100,
            2
        )
    )

percent_to_moneyline(53)
```

```python
import numpy as np
import matplotlib.pyplot as plt

_, ax = plt.subplots(1, 2, figsize=(20, 10))

x = np.linspace(10, 50, 49)
y = [ percent_to_moneyline(v) for v in x ]

ax[0].plot(x, y)
ax[0].set_title('percent < 50')

x = np.linspace(51, 90, 49)
y = [ percent_to_moneyline(v) for v in x ]

ax[1].plot(x, y)
ax[1].set_title('percent > 50')

for subplot in ax:
    subplot.set(xlabel='percent', ylabel='moneyline')

plt.show()
```
